```python
column_with_outliers = []

for col in numeric:

  if col != 'mental_health_score':

      Q1 = numeric[col].quantile(0.25)
      Q3 = numeric[col].quantile(0.75)
      IQR = Q3 - Q1
      lower = Q1 - 1.5 * IQR
      upper=  Q3 + 1.5 * IQR

      if ((numeric[col] < lower) | (numeric[col] > upper)).any:
          column_with_outliers.append(col)

for col in column_with_outliers:
  print(col)
```

```python
import numpy as np

for col in column_with_outliers:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = np.clip(df[col], lower, upper)
```

```python
X = df.drop(columns=['y'])
```

```python
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

def evaluate_clustering(X, labels, name):
    valid_labels = labels[labels != -1]
    valid_X = X[labels != -1]
    if len(set(valid_labels)) > 1:
        silhouette = silhouette_score(valid_X, valid_labels)
        dbi = davies_bouldin_score(valid_X, valid_labels)
        chi = calinski_harabasz_score(valid_X, valid_labels)
        print(f"\n{name} Results:")
        print(f"Silhouette Score: {silhouette:.4f}")
        print(f"Davies-Bouldin Index: {dbi:.4f}")
        print(f"Calinski-Harabasz Index: {chi:.2f}")
    else:
        print(f"\n{name} Results: Not enough clusters for evaluation")

kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

dbscan = DBSCAN(eps=1.5, min_samples=10)
dbscan_labels = dbscan.fit_predict(X)

agg = AgglomerativeClustering(n_clusters=4)
agg_labels = agg.fit_predict(X)

evaluate_clustering(X, kmeans_labels, "KMeans")
evaluate_clustering(X, dbscan_labels, "DBSCAN")
evaluate_clustering(X, agg_labels, "Agglomerative")
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import numpy as np
import pandas as pd

feature_columns = ['age', 'balance', 'duration', 'campaign']
X = df[feature_columns]

X_processed = X

sse = []
K = range(2, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_processed)
    sse.append(kmeans.inertia_)

plt.plot(K, sse, 'bx-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('SSE')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=42)
df[['tsne1', 'tsne2']] = tsne.fit_transform(X)

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='tsne1', y='tsne2', hue='cluster', palette='viridis')
plt.title('t-SNE - Cluster Visualization')
plt.show()
```
